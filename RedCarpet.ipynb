```python
# It has been observed that the NaN values in funding value columns are zero as no funding 
# has been raised in that year or previous year or raised a little bit as raised for the first time in next year.
df_data[['C1) Funding FY2008','C2) Funding FY2009']] = df_data[['C1) Funding FY2008','C2) Funding FY2009']].fillna(0)
```

```python
F2008 = df_data['C1) Funding FY2008']
F2009 = df_data['C2) Funding FY2009']
print(F2009)
```

```python
df_data['Growth of funding b/w 2008 & 2009'] = (F2009-F2008)*100/F2008
```

```python
df_data['Growth of funding b/w 2008 & 2009']
```

```python
def tagging(Growth):
    if Growth < 0:
        return 0
    if Growth >= 0:
        return 1
    if Growth == float("inf"):
        return 1
    
```

```python
df_data['Target'] = df_data['Growth of funding b/w 2008 & 2009'].apply(tagging)
```

```python
df_data['Target'] = df_data['Target'].fillna(0) # we got a NaN where it was zero in both funding years, 
                                                # and may be considered as negtive as compared to others
```

```python
df_data['Target'] = df_data['Target'].astype('int64')
```

```python
# Some Features are described in multicolumns. To plot count of those function is defined as:

def multi_columns(pl, c1, c2, figsize = (12, 10)):
    to_plot = [pl]

    f = {}

    for i in range(c1, c2):
        to_plot.append('Unnamed: {}'.format(i))
    
    for p in to_plot:
        freq = df_data[p].dropna().value_counts()
        f[freq.index[0]] = freq.iloc[0]

    # print(to_plot)
    f = pd.Series(f)
    plt.figure(figsize = figsize)
    plt.ylabel('Frequency')
    plt.title(to_plot[0])   
    f.plot.bar()
```

```python
multi_columns('F2) Secondary Investment Objectives', 13, 21)
```
